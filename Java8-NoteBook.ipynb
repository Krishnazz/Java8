```java
/* KeyValue pair Key - String Value - Length of the String */

import java.util.*;
ArrayList<String> al=new ArrayList<>();
		
		al.add("Sam");
		al.add("Kaira");
		al.add("Kayadu");
		al.add("Pooja");
		al.add("Rukmini");
		
		Map<String ,Integer>map=al.stream().collect(Collectors.toMap(i->i, i->i.length()));
		System.out.print(map);
```

```java
/*Map-List -- Map contains Key as String value as List */

Map<String,List<String>>map=new HashMap<>();
		map.put("Fruits", Arrays.asList("Apple","Orange","Mango"));
		map.put("Vegetables", Arrays.asList("Carrot","potato","beetroot"));
		map.forEach((key,value)->{
			System.out.println("KEY -- "+key);
			value.forEach(i->System.out.println(i));
		});
```
